```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.exp(x)
plt.plot(x, y, label='exp(x)')
plt.title('График экспоненциальной функции')
plt.xlabel('x')
plt.ylabel('exp(x)')
```

```python

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(1, 10, 100)


y = np.log(x)
y1 = np.log2(x)
y2 = np.log10(x)
plt.plot(x, y, label='ln(x)')
plt.plot(x,y1, label='log2(x)')
plt.plot(x, y2, label='log10(x)')
plt.title('Графики логарифмических функций')
plt.xlabel('x')
plt.ylabel('y')
```

```python
x = np.linspace(0, 1, 100)
y = np.arcsin(x)
y1 = np.arccos(x)
y2 = np.arctan(x)
plt.plot(x, y, 'r-', label='Арксинус')  
plt.plot(x, y1, 'g--', label='Арккосинус')
plt.plot(x, y2, 'y-.', label='Арктангенс')  
plt.title('Графики арктригонометрических функций')
plt.xlabel('x')
plt.ylabel('y')
```

```python
x = np.linspace(0, 10, 100)
y = np.sinh(x)
plt.plot(x, y)
plt.title('График гиперболического синуса')
plt.xlabel('Линейный массив')
plt.ylabel('sinh(x)')
plt.xlim(-2, 12)
plt.ylim(-1000, 12000)
```

```python
x2 = np.linspace(1, 10, 100)

# 2.2. Натуральный логарифм от x
y2 = np.log(x2)

# 2.3. Логарифм по основанию 2 от x
y12 = np.log2(x2)

# 2.4. Логарифм по основанию 10 от x
y22 = np.log10(x2)

# Построение графиков для задания 2
plt.figure(figsize=(12, 8))

# График натурального логарифма
plt.subplot(231)
plt.plot(x2, y2, label='ln(x)')
plt.title('Натуральный логарифм')
plt.xlabel('x')
plt.ylabel('ln(x)')


# График логарифма по основанию 2
plt.subplot(232)
plt.plot(x2, y12, label='log2(x)')
plt.title('Логарифм по основанию 2')
plt.xlabel('x')
plt.ylabel('log2(x)')

# График логарифма по основанию 10
plt.subplot(233)
plt.plot(x2, y22, label='log10(x)')
plt.title('Логарифм по основанию 10')
plt.xlabel('x')
plt.ylabel('log10(x)')

x3 = np.linspace(0, 1, 100)
y3 = np.arcsin(x3)
y13 = np.arccos(x3)
y23 = np.arctan(x3)

# График арксинуса
plt.subplot(234)
plt.plot(x3, y3, 'r-', label='Арксинус')
plt.title('Арксинус')
plt.xlabel('x')
plt.ylabel('arcsin(x)')


# График арккосинуса
plt.subplot(235)
plt.plot(x3, y13, 'g--', label='Арккосинус')
plt.title('Арккосинус')
plt.xlabel('x')
plt.ylabel('arccos(x)')

# График арктангенса
plt.subplot(236)
plt.plot(x3, y23, 'y-.', label='Арктангенс')
plt.title('Арктангенс')
plt.xlabel('x')
plt.ylabel('arctan(x)')

```
